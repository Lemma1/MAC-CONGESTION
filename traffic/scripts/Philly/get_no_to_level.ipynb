```python
linkid2no = dict()
for line in file('match_file', 'r'):
    words = line.split(' ')
    linkid = int(words[0])
    no_list = list()
    for e in words[1].split('+'):
        no_list.append(int(e.split(':')[0]))
    linkid2no[linkid] = no_list
```

```python
def get_typeno(l):
#     print l
    total_length = np.float(0)
    total_weight = np.float(0)
    for no in l:
        typeno = match_dict[no]
        rec = a[a['NO'] == no]
#         print rec
        length = rec.iloc[0]['LENGTH']
#         print length
        length = np.maximum(np.float(length[:-2]), 1e-6)
        total_length += length
        total_weight += length * typeno
    return int(np.round(total_weight/total_length))
```

```python
link_match_dict = dict()
for linkid in linkid2no.keys():
    if len(linkid2no[linkid]) != 0:
        typeno = get_typeno(linkid2no[linkid])
        link_match_dict[linkid] = typeno
    else:
        print linkid
    
```
